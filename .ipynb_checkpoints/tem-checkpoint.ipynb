```python
def data_process(df):
    df['stime'] = pd.to_datetime(df['stime'], format="%Y/%m/%d %H:%M:%S")
    df['year'] = df['stime'].dt.year
    df['month'] = df['stime'].dt.month
    
    train_data = df[
    (df['lon'] == 102.03) &
    (df['lat'] == 28.23)&
    (df['year'] != 2022)]
    
    test_data = test = df[
    (df['year'] == 2022) &
    (df['lon'] == 102.03) &
    (df['lat'] == 28.23)
]
    return train_data , test_data
```

```python
def scheduler(epoch, lr):
    start_decay_epoch = 100
    end_decay_epoch = 200
    decay_rate = 0.1

    if start_decay_epoch <= epoch < end_decay_epoch:
        return lr * decay_rate
    else:
        return lr
```

```python
def process_outlier(df,feats):
    _ = df[feats[7:]]
    # 1. 计算每个气象要素的平均值和标准差
    mean_values = _.mean()
    std_values = _.std()
    # 2. 根据三倍标准差法，设定异常值的阈值
    threshold = 3 * std_values
    # 3. 遍历每个气象要素的数值，将超过设定阈值的值标记为异常值
    is_outlier = (_ > mean_values + threshold) | (_ < mean_values - threshold)
    # 4. 对于标记为异常值的数据，可以根据需要选择删除、替换或进行插补处理
    # 假设你选择删除异常值
    df = df[~is_outlier.any(axis=1)]
#     df = df.query('sw_max1<20 and sw_max2<20')
     
    return df
```

```python
# 时间特征 *   
def get_time_features(df):
#     df['year'] = df.time.dt.year
    df['month'] = df.time.dt.month
    df['day'] = df.time.dt.day
    df['hour'] = df.time.dt.hour
    return df
```

```python
# 构造窗口统计特征 *
def get_window(df,ori_feats):
    df=df.sort_values(['stime','dtime']).reset_index(drop=True) 
    
    for feat in ori_feats:
        # window=8的原因是8为一个周期
        df[feat+"_max_8"] = df.groupby('stime')[feat].rolling(window=8).max().reset_index(drop=True)
        df[feat+"_avg_8"] = df.groupby('stime')[feat].rolling(window=8).mean().reset_index(drop=True)
        df[feat+"_var_8"] = df.groupby('stime')[feat].rolling(window=8).var().reset_index(drop=True)
        df[feat+"_min_8"] = df.groupby('stime')[feat].rolling(window=8).min().reset_index(drop=True)


    df = df.fillna(method='bfill')
    df = df.fillna(method='ffill')
    
    return df
```

```python
# 构造时滞特征 
def get_lag_features(df,ori_feats):
    df=df.sort_values(['stime','dtime']).reset_index(drop=True)
    
    
    for feat in ori_feats:
        df[feat+'_s1'] = df.groupby('stime')[feat].shift(1)
        
    # 构造后可能首尾出现空值
    df = df.fillna(method='bfill')
    df = df.fillna(method='ffill')

    return df
```

```python
def select_features(df,thus):
    
    df_corr = df.drop_duplicates(subset=['time'])
    df_corr_1_res = pd.DataFrame(df_corr.corr()['tp'])
    df_corr_1_res.sort_values('tp',ascending=False)
    df_corr_1_res.to_csv('tp_corr.csv')
    
    df_corr_1_res = df_corr_1_res.query('abs(tp)>=@thus')
    
    feats = list(set(df_corr_1_res.index))
    feats = list(filter(lambda x: x not in ['tp','time','stime','time_order'] , feats))
    print(f'corr>{thus}:{feats}')
    
    feats_name = ['level', 'time', 'dtime','stime', 'id', 'lon', 'lat','time_order','tp']+feats
    return feats_name
```

```python
num_cols = x_train.shape[1]
# 构建神经网络模型
strategy = tf.distribute.MirroredStrategy()

# 打印可用 GPU 数量
print('Number of devices: {}'.format(strategy.num_replicas_in_sync))

with strategy.scope():
    model = Sequential([

        Dense(512, activation='tanh', input_shape=(num_cols,)),  # 输入形状为10
        #Dropout(0.2), # 20% 的神经元在每次训练迭代中被置为 0
        Dense(256,activation='sigmoid'),
        # Dropout(0.2), # 20% 的神经元在每次训练迭代中被置为 0
        Dense(128,activation='tanh'),
        Dense(64, activation='sigmoid'),
        Dense(32,activation='tanh'),
        Dense(16,activation='tanh'),
        Dense(8,activation='tanh'),
        Dense(1, activation='linear')
    ])


# 定义RMSE损失函数
def rmse_loss(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))


def quantile_loss(y_true, y_pred):
    q = 0.8  # 可以根据你的需要修改
    error = y_true - y_pred
    return tf.reduce_mean(tf.maximum(q * error, (q - 1) * error))


#Adam优化器
optimizer = Adam(lr=0.001)
#随机梯度下降

model.compile(optimizer=optimizer, loss=quantile_loss, metrics=['mse'])

#早停法：设置20轮
# 创建EarlyStopping回调
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)


# 创建 LearningRateScheduler 回调
lr_scheduler = LearningRateScheduler(scheduler)

# 将 lr_scheduler 添加到回调列表中
callbacks = [early_stopping,lr_scheduler]

# 在 fit() 函数中使用 callbacks 参数
history = model.fit(x_train, y_train, epochs=2000, validation_split=0.2, callbacks=callbacks)

test_loss, test_mae = model.evaluate(x_test, y_test)


# 测试集上的平均绝对误差
print("Test Mean Absolute Error (MAE): {:.2f}".format(test_mae))

# 模型保存的路径和文件名
model_tem = "/home/lihaifei/yuchen/046/temp.h5"


model.save(model_tem)
```

```python
def preprocess_data(data, scaler_path,feats_name):
    features = data[feats_name]
    features['time'] = pd.to_datetime(features['time'])
    features['time'] = features['time'].apply(lambda x: x.timestamp())
    features['stime'] = pd.to_datetime(features['time'])
    features['stime'] = features['stime'].apply(lambda x: x.timestamp())
    scaler = joblib.load(scaler_path)
    features_scaled = scaler.transform(features)
    
    return features_scaled
```

```python
filtered_predictions_df = pd.DataFrame()

# 对每个起始时间应用卡尔曼滤波器
for stime, group in predictions_df.groupby('stime'):
    # 按照dtime排序，保证预报按照时间顺序
    group = group.sort_values('dtime')

    predicted_temp = group['tem_neural_network'].values  # 获取当前时间序列的神经网络预测的温度值

    initial_state = predicted_temp[0]  # 初始状态设为神经网络预测的第一个温度值
    filtered_state_means = np.zeros(predicted_temp.shape)  # 初始化一个用于存放滤波状态均值的数组
    filtered_state_covariances = np.zeros(predicted_temp.shape)  # 初始化一个用于存放滤波状态协方差的数组
    global_error_variance = np.var(predicted_temp - filtered_state_means)  # 计算一个全局误差方差

    kf = KalmanFilter(transition_matrices=[1],
                      observation_matrices=[1],
                      initial_state_mean=initial_state,
                      initial_state_covariance=1,
                      transition_covariance=np.var(predicted_temp))

    for t in range(len(predicted_temp)):
        if t == 0:
            # 第一个点，用神经网络初始预测值
            filtered_state_means[t] = initial_state
            filtered_state_covariances[t] = 1
        else:
            # 更新卡尔曼滤波
            filtered_state_means[t], filtered_state_covariances[t] = kf.filter_update(
                filtered_state_mean=filtered_state_means[t-1],
                filtered_state_covariance=filtered_state_covariances[t-1],
                observation=predicted_temp[t],
                observation_matrix=np.atleast_2d(1),
                observation_covariance=0.5*global_error_variance
            )

    # 将处理后的预报添加到新的DataFrame
    group['tem_kalman_filter'] = filtered_state_means
    filtered_predictions_df = filtered_predictions_df.append(group)


filtered_predictions_df['ensemble'] = (filtered_predictions_df['tem_neural_network'] + filtered_predictions_df['tem_kalman_filter']) / 2


# 保存包含滤波后预测的新DataFrame
filtered_predictions_df.to_csv('/share/HPC/home/zhaox_hx/data/aidata/zhangym_tmp/Project/046_project/code/data/预报/new_dataset_yyc/predictions1.csv', index=False)
```
